```python
def contains_n(text):
    if 'N' in text:
        return True
    else:
        return False

# Or the shorter, more pythonic way
def contains_n(text):
    bool_val = 'N' in text
    return bool_val
```

```python
df['On_N_Line'] = df['LINENAME'].map(contains_n)
df.head(2)
```
