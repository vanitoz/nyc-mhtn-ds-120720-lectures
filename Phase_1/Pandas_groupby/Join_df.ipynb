```python
cards_df = pd.read_csv('cards.csv')
dust_df = pd.read_csv('dust.csv')
entourages_df = pd.read_csv('entourages.csv')
mechanics_df = pd.read_csv('mechanics.csv')
play_requirements_df = pd.read_csv('play_requirements.csv')
```

```python
cards_df.set_index(cards_df['card_id'], inplace= True)
cards_df.drop(['card_id'],axis=1,inplace=True)
dust_df.set_index(dust_df['card_id'], inplace= True)
dust_df.drop(['card_id'],axis=1,inplace=True)
entourages_df.set_index(entourages_df['card_id'], inplace= True)
entourages_df.drop(['card_id'],axis=1,inplace=True)
mechanics_df.set_index(mechanics_df['card_id'], inplace= True)
mechanics_df.drop(['card_id'],axis=1,inplace=True)
play_requirements_df.set_index(play_requirements_df['card_id'], inplace= True)
play_requirements_df.drop(['card_id'],axis=1,inplace=True)
```
